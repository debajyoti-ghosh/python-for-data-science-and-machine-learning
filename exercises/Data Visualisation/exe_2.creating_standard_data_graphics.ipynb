# Exercises: Practice Data Visualization

Use the `mtcars` dataset and the skills you've learned to complete the following exercises. Try to write your own code for each task, then compare with the suggested solutions.

## Exercise 1: Line Charts
1. Create a line chart for the `hp` (horsepower) column from the `mtcars` dataset.
2. Plot a line chart for both `disp` and `wt` columns in the same figure.
3. Customize the plot with a title, X and Y labels, and a legend.

## Exercise 2: Bar Charts
1. Create a vertical bar chart for the `gear` column.
2. Create a horizontal bar chart for the `carb` column.
3. Compare `mpg` values for the first 5 cars using a vertical bar chart with custom colors.

## Exercise 3: Pie Charts
1. Create a pie chart showing the proportion of cars with different numbers of cylinders (`cyl` column).
2. Add labels and percentages to the pie chart.
3. Save the pie chart as an image called `cyl_pie_chart.png`.

## Exercise 4: Custom Data
1. Generate a random dataset with 10 values using `np.random.randn()`.
2. Create a line chart, a bar chart, and a pie chart from this random data.
3. Save all plots as images with meaningful filenames.

## Exercise 5: Bonus Challenge
1. Using Matplotlib, combine multiple types of plots in a single figure:
   - A line chart for `mpg`
   - A bar chart for `hp`
   - Ensure each chart is clearly labeled and distinguishable.
2. Experiment with colors, markers, and line styles to improve visualization clarity.
